```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext,PromptTemplate
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
document=SimpleDirectoryReader("/content/sample_data/pdfdata").load_data()
document
```

```python
from llama_index.core import PromptTemplate
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as accurately as possible
based on the instructions and context provided.
"""
qa_template = PromptTemplate(system_prompt)
messages = qa_template.format_messages(context_str=..., query_str=...)
```

```python
!huggingface-cli login
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=qa_template,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    model_kwargs={"torch_dtype": torch.float16}
)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding

embedding_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
)
```

```python
from llama_index.core import Settings
from llama_index.core.node_parser import SentenceSplitter

Settings.llm = llm
Settings.embed_model = embedding_model
Settings.node_parser = SentenceSplitter(chunk_size=1024)
```

```python
# a vector store index only needs an embed model
index = VectorStoreIndex.from_documents(document, embed_model=embedding_model)

query_engine = index.as_query_engine(llm=llm)
```

```python
response = query_engine.query('Who is Sushmitha?')
```
